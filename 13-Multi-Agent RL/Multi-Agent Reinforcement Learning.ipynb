## Problem 1: Nash Equilibrium (Theory)

A Nash Equilibrium (NE) represents a state where no player can improve their outcome by unilaterally changing their strategy. For our games, we'll focus on finding the mixed-strategy NE, where players choose their actions probabilistically.

### 1.1 Standard Rock-Scissors-Paper

Given the standard RSP payoff matrix:

| Player 1 | Rock | Scissors | Paper |
| :--- | :--: | :---: | :---: |
| **Rock** | 0, 0 | 1, -1 | -1, 1 |
| **Scissors**| -1, 1 | 0, 0 | 1, -1 |
| **Paper** | 1, -1 | -1, 1 | 0, 0 |


**Your Task:** Analytically derive the mixed-strategy Nash Equilibrium for this game. Show the steps for setting up the indifference equations for Player 1 and solving for Player 2's equilibrium strategy probabilities $(q_R, q_S, q_P)$. (find the Mixed Nash equilibrium of the game)

## Solution — Mixed-Strategy Nash Equilibrium for Standard RSP

Let Player 2 play Rock/Scissors/Paper with probabilities $(q_R, q_S, q_P)$ (with $q_R+q_S+q_P=1$).
Player 1’s expected payoff from each pure action against $(q_R,q_S,q_P)$ is:

* If P1 plays **Rock**: $\;u_1(R)=0\cdot q_R + 1\cdot q_S + (-1)\cdot q_P = q_S - q_P.$
* If P1 plays **Scissors**: $\;u_1(S)=(-1)\cdot q_R + 0\cdot q_S + 1\cdot q_P = q_P - q_R.$
* If P1 plays **Paper**: $\;u_1(P)=1\cdot q_R + (-1)\cdot q_S + 0\cdot q_P = q_R - q_S.$

At a mixed-strategy Nash equilibrium, Player 1 must be **indifferent** among her pure strategies (otherwise she would deviate). Hence,

$$
u_1(R) = u_1(S) = u_1(P).
$$

Equating pairwise:

$$
\begin{aligned}
q_S - q_P &= q_P - q_R \quad\Rightarrow\quad q_R + q_S = 2q_P \quad\quad (1)\\
q_P - q_R &= q_R - q_S \quad\Rightarrow\quad q_P + q_S = 2q_R \quad\quad (2)
\end{aligned}
$$

together with the probability constraint

$$
q_R + q_S + q_P = 1. \quad\quad (3)
$$

Solve (2) with (3): from (2), $q_P+q_S=2q_R$. Plug into (3):

$$
2q_R + q_R = 1 \;\Rightarrow\; 3q_R = 1 \;\Rightarrow\; q_R = \tfrac{1}{3}.
$$

Then $q_P + q_S = \tfrac{2}{3}$. Use (1): $q_R + q_S = 2q_P \Rightarrow \tfrac{1}{3} + q_S = 2q_P$.
Substitute $q_P = \tfrac{2}{3} - q_S$:

$$
\tfrac{1}{3} + q_S = 2\!\left(\tfrac{2}{3} - q_S\right) = \tfrac{4}{3} - 2q_S
\;\Rightarrow\; 3q_S = 1 \;\Rightarrow\; q_S = \tfrac{1}{3},
$$

and hence $q_P = \tfrac{2}{3} - \tfrac{1}{3} = \tfrac{1}{3}$.

Therefore, Player 2’s equilibrium mixed strategy is

$$
(q_R, q_S, q_P) = \left(\tfrac{1}{3}, \tfrac{1}{3}, \tfrac{1}{3}\right).
$$

By symmetry of the zero-sum RSP game, Player 1’s equilibrium mix is the same:

$$
(p_R, p_S, p_P) = \left(\tfrac{1}{3}, \tfrac{1}{3}, \tfrac{1}{3}\right),
$$

and the game value is $0$ (each player’s expected payoff is $0$ at equilibrium).


### 1.2 Modified Rock-Scissors-Paper

Now, consider the modified RSP game where the stakes are higher:

| Player 1 | Rock | Scissors | Paper |
| :--- | :--: | :---: | :---: |
| **Rock** | 0, 0 | 1, -1 | -2, 2 |
| **Scissors**| -1, 1 | 0, 0 | 3, -3 |
| **Paper** | 2, -2 | -3, 3 | 0, 0 |


**Your Task:** Like pervious one Derive the mixed-strategy Nash Equilibrium for this modified game.

## Solution — Mixed-Strategy Nash Equilibrium for Modified RSP

Let Player 2 mix over **Rock/Scissors/Paper** with probabilities $(q_R,q_S,q_P)$ (so $q_R+q_S+q_P=1$).
Player 1’s expected payoff from each pure action (using the given payoff matrix for Player 1) is:

* If P1 plays **Rock**: $u_1(R)=0\cdot q_R + 1\cdot q_S + (-2)\cdot q_P = q_S - 2q_P.$
* If P1 plays **Scissors**: $u_1(S)=(-1)\cdot q_R + 0\cdot q_S + 3\cdot q_P = -q_R + 3q_P.$
* If P1 plays **Paper**: $u_1(P)=2\cdot q_R + (-3)\cdot q_S + 0\cdot q_P = 2q_R - 3q_S.$

At a mixed-strategy NE, Player 1 must be **indifferent** among her pure actions that receive positive probability. Hence,

$$
u_1(R)=u_1(S)=u_1(P).
$$

Equate pairwise and use the probability constraint:

$$
\begin{aligned}
q_S - 2q_P &= -q_R + 3q_P \quad\Rightarrow\quad q_R + q_S = 5q_P \quad (1)\\
-q_R + 3q_P &= 2q_R - 3q_S \quad\Rightarrow\quad q_P + q_S = q_R \quad (2)\\
q_R + q_S + q_P &= 1. \quad (3)
\end{aligned}
$$

From (2), $q_R = q_P + q_S$. Substitute into (1):

$$
(q_P + q_S) + q_S = 5q_P \;\Rightarrow\; 2q_S = 4q_P \;\Rightarrow\; q_S = 2q_P.
$$

Then $q_R = q_P + q_S = 3q_P$. Using (3):

$$
q_R + q_S + q_P = 3q_P + 2q_P + q_P = 6q_P = 1 \;\Rightarrow\; q_P=\tfrac{1}{6}.
$$

Therefore,

$$
(q_R,q_S,q_P)=\left(\tfrac{1}{2},\,\tfrac{1}{3},\,\tfrac{1}{6}\right).
$$

**Verification / value:**
Plugging into any payoff:

$$
u_1(R)=\tfrac{1}{3}-2\cdot\tfrac{1}{6}=0,\quad
u_1(S)=-\tfrac{1}{2}+3\cdot\tfrac{1}{6}=0,\quad
u_1(P)=2\cdot\tfrac{1}{2}-3\cdot\tfrac{1}{3}=0.
$$

So the game value is $0$. By the same indifference logic for Player 2 (or noting the matrix is skew-symmetric), Player 1’s equilibrium mix is identical:

$$
(p_R,p_S,p_P)=\left(\tfrac{1}{2},\,\tfrac{1}{3},\,\tfrac{1}{6}\right).
$$

## Problem 2: Learning by Observation - Fictitious Play (Implementation)

Fictitious Play is an intuitive learning algorithm where each agent models its opponent as playing a stationary strategy defined by the historical frequency of their past actions. The agent then plays a **best response** to this belief.

### 2.1 Implementation

**Your Task:** Implement the `simulate_fictitious_play` function below. It should take the payoff matrices for both players and the number of iterations as input. At each step, each player should choose the action that maximizes their expected payoff given the history of the opponent's plays.

**Algorithm:** At each time step $t > 0$, Player $i$ forms a belief that their opponent ($-i$) will play each action $a'$ with a probability equal to its historical frequency. The agent then chooses an action $a_i^*$ that is a best response to this belief.

Let $C_{t-1}(a_{-i})$ be the count of times opponent $-i$ has played action $a_{-i}$ up to step $t-1$. Player $i$'s best response is:
$$a_{i,t}^* = \arg\max_{a_i \in A_i} \sum_{a_{-i} \in A_{-i}} u_i(a_i, a_{-i}) \cdot \frac{C_{t-1}(a_{-i})}{t-1}$$

**Note on Tie-Breaking:** If multiple actions yield the same maximal expected payoff, your agent should choose one of these best responses uniformly at random.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulate_fictitious_play(A, B, iterations):
    """
    Simulates Fictitious Play for two players in a normal-form game.

    Args:
        A (np.ndarray): Payoff matrix for Player 1 of shape (m, n).
        B (np.ndarray): Payoff matrix for Player 2 of shape (m, n).
        iterations (int): The number of rounds to play.

    Returns:
        tuple: A tuple containing:
            - p1_freq_history (np.ndarray): shape (iterations, m), history of Player 1's action frequencies.
            - p2_freq_history (np.ndarray): shape (iterations, n), history of Player 2's action frequencies.
    """
    m, n = A.shape
    assert B.shape == (m, n), "A and B must have the same shape"

    # Action counts
    p1_counts = np.zeros(m, dtype=int)
    p2_counts = np.zeros(n, dtype=int)

    # History of empirical frequencies (for plotting/analysis)
    p1_freq_history = np.zeros((iterations, m), dtype=float)
    p2_freq_history = np.zeros((iterations, n), dtype=float)

    rng = np.random.default_rng()

    for t in range(iterations):
        if t == 0:
            # Tie-break for the very first move: pick action 0 for both players
            a1 = 0
            a2 = 0
        else:
            # Beliefs: opponent's empirical frequencies up to t-1
            p2_belief = p2_counts / t  # length n
            p1_belief = p1_counts / t  # length m

            # Expected payoffs for each action under the belief
            # P1's expected payoff for each action i: (A @ p2_belief)[i]
            p1_expected = A @ p2_belief  # shape (m,)
            # P2's expected payoff for each action j: (B.T @ p1_belief)[j]
            p2_expected = B.T @ p1_belief  # shape (n,)

            # Best responses with uniform random tie-breaking
            max1 = np.max(p1_expected)
            best1 = np.flatnonzero(np.isclose(p1_expected, max1))
            a1 = rng.choice(best1)

            max2 = np.max(p2_expected)
            best2 = np.flatnonzero(np.isclose(p2_expected, max2))
            a2 = rng.choice(best2)

        # Play actions and update counts
        p1_counts[a1] += 1
        p2_counts[a2] += 1

        # Record current frequencies after round t
        p1_freq_history[t] = p1_counts / (t + 1)
        p2_freq_history[t] = p2_counts / (t + 1)

    return p1_freq_history, p2_freq_history

# --- Payoff Matrices ---
A_std = np.array([[0, 1, -1], [-1, 0, 1], [1, -1, 0]])
A_mod = np.array([[0, 1, -2], [-1, 0, 3], [2, -3, 0]])
```

## Problem 3: Fictitious Play with Exploration (Implementation)

Our Fictitious Play agent is purely exploitative. In Reinforcement Learning, we know the importance of the **exploration-exploitation tradeoff**. Let's create an $\epsilon$-greedy version of Fictitious Play.

### 3.1 Implementation

**Your Task:** Create a new function, `simulate_epsilon_greedy_fp`. This function should be similar to your Fictitious Play implementation but include an `epsilon` parameter. At each step, with probability `epsilon`, the agent should choose a random action (explore). With probability `1-epsilon`, it should play the best response (exploit).

```python
def simulate_epsilon_greedy_fp(A, B, iterations, epsilon):
    """
    Simulates epsilon-greedy Fictitious Play for two players in a normal-form game.

    Args:
        A (np.ndarray): Payoff matrix for Player 1 of shape (m, n).
        B (np.ndarray): Payoff matrix for Player 2 of shape (m, n).
        iterations (int): Number of rounds to play.
        epsilon (float): Exploration probability in [0, 1]. With prob epsilon,
                         each player chooses a random action; otherwise best response.

    Returns:
        tuple:
            - p1_freq_history (np.ndarray): shape (iterations, m), history of P1 action frequencies.
            - p2_freq_history (np.ndarray): shape (iterations, n), history of P2 action frequencies.
    """
    m, n = A.shape
    assert B.shape == (m, n), "A and B must have the same shape"
    assert 0.0 <= epsilon <= 1.0, "epsilon must be in [0, 1]"

    rng = np.random.default_rng()

    # Action counts and frequency history
    p1_counts = np.zeros(m, dtype=int)
    p2_counts = np.zeros(n, dtype=int)
    p1_freq_history = np.zeros((iterations, m), dtype=float)
    p2_freq_history = np.zeros((iterations, n), dtype=float)

    for t in range(iterations):
        # Choose P1 action
        if rng.random() < epsilon:
            # Explore
            a1 = rng.integers(m)
        else:
            if t == 0:
                # No belief yet; follow the same tie-break rule as baseline FP
                a1 = 0
            else:
                # Exploit: best response to opponent's empirical frequencies
                p2_belief = p2_counts / t  # length n
                p1_expected = A @ p2_belief  # shape (m,)
                max1 = np.max(p1_expected)
                best1 = np.flatnonzero(np.isclose(p1_expected, max1))
                a1 = rng.choice(best1)

        # Choose P2 action
        if rng.random() < epsilon:
            a2 = rng.integers(n)
        else:
            if t == 0:
                a2 = 0
            else:
                p1_belief = p1_counts / t  # length m
                p2_expected = B.T @ p1_belief  # shape (n,)
                max2 = np.max(p2_expected)
                best2 = np.flatnonzero(np.isclose(p2_expected, max2))
                a2 = rng.choice(best2)

        # Update counts
        p1_counts[a1] += 1
        p2_counts[a2] += 1

        # Record empirical frequencies after round t
        p1_freq_history[t] = p1_counts / (t + 1)
        p2_freq_history[t] = p2_counts / (t + 1)

    return p1_freq_history, p2_freq_history
```

### Why does this happen? (Key insight)

A naïve guess is that constant exploration would pull averages toward the **uniform** distribution $u=(\tfrac{1}{3},\tfrac{1}{3},\tfrac{1}{3})$, yielding a mixture $(1-\epsilon)p^\star + \epsilon u$.
**However**, in this zero-sum game the NE has **full support**: against the opponent’s NE, **all three pure actions are payoff-equivalent**. That means any mixture over $\{R,S,P\}$ is a best response at the equilibrium belief.

Because exploration injects uniform random play a fraction $\epsilon$ of the time, the **exploit** part (the best-response actions chosen with probability $1-\epsilon$) compensates in the opposite direction. In the long run, the exploited action frequencies adapt to

$$
e(\epsilon)\;=\;\frac{p^\star-\epsilon u}{1-\epsilon},
$$

so that the **total** empirical mix
$(1-\epsilon)\,e(\epsilon)\;+\;\epsilon\,u \;=\; p^\star.$

This compensation is **feasible** as long as every coordinate of $e(\epsilon)$ is nonnegative, i.e. $\epsilon \le 3\min_i p^\star_i$.
Here $\min_i p^\star_i = 1/6$, so the bound is $\epsilon \le 0.5$.
Your runs ($\epsilon=0.01,0.1,0.3$) satisfy this, hence convergence **to the same NE**.

### Takeaways

* **Effect of ε:** Larger $\epsilon$ adds variance (noisier curves, slightly slower smoothing) but, for moderate $\epsilon$, **does not bias** the long-run averages away from NE in this game.
* **Convergence target:** With $\epsilon \le 0.5$ here, the **empirical strategies converge to the NE**.
* **When would ε move the limit?** If $\epsilon$ were too large (e.g., $>0.5$ in this game), the compensating exploit mix $e(\epsilon)$ would become infeasible (some component negative), and the averages would be pulled away from $p^\star$.
* **Broader note:** This result leverages the **zero-sum** structure and **full-support** NE. In general-sum games or games with **non-interior** NE, constant exploration can shift the limit away from the NE.

## Problem 4: Learning from "What If" - Regret Matching (Implementation & Theory)

Regret Matching is a powerful no-regret learning algorithm. Instead of playing a best response to history, an agent's probability of choosing an action is proportional to the positive **regret** for not having chosen that action in the past. The key property of regret matching is that the **average strategy** over time converges to a Nash Equilibrium.

### 4.1 Implementation

**Your Task:** Implement the `simulate_regret_matching` function below.

**Algorithm:** Regret Matching works in two steps. First, update the cumulative regrets. Second, determine the next round's strategy.

1.  **Regret Calculation:** After playing action $a_i$ against opponent's action $a_{-i}$, the cumulative regret $R_t(s)$ for *not* having played action $s \in A_i$ is updated as follows:
    $$R_t(s) = R_{t-1}(s) + u_i(s, a_{-i}) - u_i(a_i, a_{-i})$$

2.  **Strategy Calculation:** The probability of playing action $s$ in the next round is proportional to its positive cumulative regret, $R_t^+(s) = \max(0, R_t(s))$.
    $$p_{t+1}(s) = \frac{R_t^+(s)}{\sum_{s' \in A_i} R_t^+(s')}$$
    If the sum of positive regrets is zero, play uniformly at random.

```python
def simulate_regret_matching(A, B, iterations):
    """
    Simulates Regret Matching for two players in a normal-form game.

    Args:
        A (np.ndarray): Payoff matrix for Player 1 of shape (m, n).
        B (np.ndarray): Payoff matrix for Player 2 of shape (m, n).
        iterations (int): Number of rounds to play.

    Returns:
        tuple:
            - p1_avg_strat_hist (np.ndarray): shape (iterations, m), history of Player 1's average strategy.
            - p1_inst_strat_hist (np.ndarray): shape (iterations, m), history of Player 1's instantaneous strategy.
    """
    m, n = A.shape
    assert B.shape == (m, n), "A and B must have the same shape"

    rng = np.random.default_rng()

    # Cumulative regrets (can be negative; we use positive part to form strategies)
    R1 = np.zeros(m, dtype=float)
    R2 = np.zeros(n, dtype=float)

    # Strategy sums to compute time-average of instantaneous strategies
    p1_strat_sum = np.zeros(m, dtype=float)
    p2_strat_sum = np.zeros(n, dtype=float)

    # Histories (we record every iteration)
    p1_inst_strat_hist = np.zeros((iterations, m), dtype=float)
    p1_avg_strat_hist  = np.zeros((iterations, m), dtype=float)

    for t in range(iterations):
        # Instantaneous mixed strategies from positive regrets
        pos1 = np.maximum(R1, 0.0)
        if pos1.sum() > 0:
            sigma1 = pos1 / pos1.sum()
        else:
            sigma1 = np.full(m, 1.0 / m)

        pos2 = np.maximum(R2, 0.0)
        if pos2.sum() > 0:
            sigma2 = pos2 / pos2.sum()
        else:
            sigma2 = np.full(n, 1.0 / n)

        # Record instantaneous strategy and update average (of instantaneous mixes)
        p1_inst_strat_hist[t] = sigma1
        p1_strat_sum += sigma1
        p2_strat_sum += sigma2
        p1_avg_strat_hist[t] = p1_strat_sum / (t + 1)

        # Sample actions for both players
        a1 = rng.choice(m, p=sigma1)
        a2 = rng.choice(n, p=sigma2)

        # Update cumulative regrets for ALL actions
        # Player 1 regrets: A[:, a2] - A[a1, a2]
        R1 += A[:, a2] - A[a1, a2]
        # Player 2 regrets: B[a1, :] - B[a1, a2]
        R2 += B[a1, :] - B[a1, a2]

    return p1_avg_strat_hist, p1_inst_strat_hist
```

### What do we observe?

* The **instantaneous strategy** continues to **oscillate** and is **far from NE** at the final step (L1 distance $\approx 1.78\times 10^{-1}$). This cycling is typical: regrets react to realized outcomes and keep shifting probability mass.
* The **average strategy** is **very close** to the NE (L1 distance $\approx 2.6\times 10^{-3}$). In the plot, its three curves track and stabilize near the dotted NE lines.

### Which one converges to the NE?

* The **average strategy** converges to the NE; the **instantaneous strategy** generally does **not**.

### Why is this expected (theory sketch)?

* **Regret Matching** (Hart & Mas-Colell) guarantees **vanishing external regret**: the per-round regret $R_t/t \to 0$.
* When **both players** minimize external regret, the **empirical joint distribution** of play converges to the set of **correlated equilibria**.
* In **two-player zero-sum** games, every correlated equilibrium yields the **minimax value**, and the **marginal averages** of each player converge to **minimax/Nash strategies**.
* Hence, $\bar{\sigma}_t \to p^\star$ (NE), while $\sigma_t$ can keep fluctuating due to the adaptive regret updates.

**Takeaway:** Regret Matching drives the **time-averaged** behavior to NE in zero-sum games; the **instantaneous** policy need not (and typically does not) converge.
